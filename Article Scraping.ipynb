```python
import numpy as np
import pandas as pd 
import requests
from bs4 import BeautifulSoup
import re
```

```python
url="https://www.karar.com/yazarlar/zafer-acar"
r=requests.get(url)
```

```python
r.status_code
```

```python
soup=BeautifulSoup(r.content,"lxml")
```

```python
page_count=len(soup.find_all("ul",{"class":"item-container"})[0].find_all("li"))-1
```

```python
page_count
```

```python
article_all=""

for page in range(1,page_count+1):
    pageRequest=requests.get(str(url)+"/?page={}".format(page))
    #print(pageRequest.url)
    pageSoup=BeautifulSoup(pageRequest.content,"lxml")
    
    entrys=pageSoup.find_all("article",{"class":"item box-shadow"})
    for entry in entrys:
        entryLink="https://www.karar.com"+entry.a.get("href")
        #print(entryLink)
        entryName=entry.a.get("title")
        entryR=requests.get(entryLink)
        article_soup=BeautifulSoup(entryR.content,"lxml")
        
        title=article_soup.find("h1",{"class":"content-title"}).text

        article=article_soup.find("article",{"class":"article-content"}).text
        
        article_all+=title+article
        #article_list.append(title)
        #article_list.append(article)       
```

```python
article_all=re.sub("[^\w\s]","",article_all)
article_all=re.sub("\n"," ",article_all)
article_all=re.sub("\r","",article_all)
article_all=article_all.lower()
```
